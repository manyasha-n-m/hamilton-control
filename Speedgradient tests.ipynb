```python
import numpy as np
from optimize import speedgrad
from models import Model1
```

```python
init_state = np.array([6,4,0,0])
```

```python
m1 = Model1(init_state)
gamma = 0.4
u = 0.1

for i in range(990):
    m1.update(u)
    u = speedgrad(m1.state, u, 8, gamma)
    
print('current state:', m1.state)
m1.plot_evolution(f'$\gamma$={gamma}')
```
