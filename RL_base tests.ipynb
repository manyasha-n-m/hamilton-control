```python
import numpy as np
from optimize import RLnetwork
from models import Model1

%matplotlib inline
```

```python
init_state = np.array([6, 4, 0, 0])
```

## standard parameters

```python
m2 = Model1(init_state)
net = RLnetwork(m2.f, m2.g, 8)
u = net.u(m2.state)

for i in range(990):
    m2.update(u)
    u = net.optimize(m2.state)

print('current state:', m2.state)
print('V(x0) =', net.V(init_state))

m2.plot_evolution(f'$X_0$={list(m2.x_0)}   $I_1^*$={net.I_[0]} real-time train')
```
