```python
import numpy as np
from time import time
from optimize import RLnetwork
from models import Model1
```

```python
init_state = np.array([6, 4, 0, 0])
```

## na1=10 na2=1

```python
m2 = Model1(init_state)
net = RLnetwork(m2.f, m2.g, 8, na1=10, na2=1)
u = net.u(m2.state)

for i in range(990):
    m2.update(u)
    t = time()
    u = net.optimize(m2.state)
    print(time()-t) if i%100==0 else _

print('current state:', m2.state)
print('V(x0) =', net.V(init_state))

m2.plot_evolution(f'N={net.N} nc1={net.nc1} nc2={net.nc2} na1={net.na1} na2={net.na2}')
```
